```python
import subprocess

def exec_command(command):
    """
    Executes a shell command.
    
    :param command: Command to be executed (string)
    :return: Output of the command execution (string)
    """
    try:
        result = subprocess.run(command, shell=True, capture_output=True, text=True)
        return result.stdout if result.returncode == 0 else result.stderr
    except Exception as e:
        return f"Error executing command: {e}"

def load_file(file_path):
    """
    Loads the content of a file.
    
    :param file_path: Path to the file (string)
    :return: Content of the file (string)
    """
    try:
        with open(file_path, 'r') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        return f"Error: File {file_path} not found."
    except Exception as e:
        return f"Error loading file: {e}"



# playwright codegen -o test.dev
```

```python
def extract_page_lines(file_content: str) -> str:
    # Split the content into lines
    lines = file_content.splitlines()

    # Extract lines that contain "page."
    extracted_lines = [line for line in lines if "page." in line]

    # Join the extracted lines back into a string
    return "\n".join(extracted_lines)
```
